# 3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

```python
class Counter:
    def __init__(self,count,step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return '[Count (step: %d)]%d' %(self.step, self.count)

    def incr(self):
        self.count += self.step
    def __call__(self, *args):
        self.incr()
        return self.__str__()
    def __add__(self,B):
        self.count += B
        return self.__str__()
    def __sub__(self, B):
        self.count -=B
        return self.__str__()
    def __gt__(self, y):
        return self.count > y

    def __ge__(self, y):
        return self.count >= y

    def __lt__(self, y):
        return self.count < y

    def __le__(self, y):
        return self.count <= y

    def __eq__(self, y):
        return self.count == y

    def __ne__(self, y):
        return self.count != y
c = Counter(10)
d = Counter(10,2)
c.incr()
d.incr()
c()
d()
c = c+5
d = d-5
print(c)
print(d)
print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

# 4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

__init__ 함수는 먼저 리스트를 받고 리스트의 원소들을 self에 append를 시켜준 다음 eliminate_duplicate()함수를 호출하여 self리스트 안의 원소들이 중복되지 않게 해주는 함수 입니다.
__str__ 함수는 self리스트 안의 원소들을 문자열로 바꿔서 문자열을 return해주는 함수이다.
eliminate_duplicate() 함수는 먼저 새로운 리스트를 선언하고 그 리스트에 인자로 받은 리스트안의 공통된 원소를 제거한 원소들을 append시켜준 다음 받은 인자 리스트를 초기화 시키고 새로운 리스트의 원소들을 append시켜주는 함수입니다.

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def __init__(self, *url):
        self.url = list()
        self.index = 0
        for e in url:
            self.url.append(e)
        
    def getKey(self, item):
        return item[1]
    def getWordsFrequency(self, reverse = False):
        mydict = self.getDict()
        keys = mydict.keys()
        values = mydict.values()
        self.mylist = list(zip(keys, values))
        if reverse == False:
            self.mylist=sorted(self.mylist, key = self.getKey, reverse = True)
        else:
            self.mylist=sorted(self.mylist, key = self.getKey, reverse = False)
        return self.mylist
    def __iter__(self): 
        return self
    def __next__(self):
        try:
            value = self.mylist[self.index]
            self.index += 1 
            return value
        except:
            raise StopIteration
    def next(self):
        return self.__next__()
```

# 2. 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.
- 인자로 전달하는 문자열은 경로만 들어간다고 가정한다.
- 각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
- 반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
- div_path('/usr/local/bin/python')
- ['/usr/local/bin', 'python']
- div_path('/home/chulsoo/test.txt')
- ['/home/chulsoo', 'test.txt']
- [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.

```python
def div_path(s):
    list = []
    a = []
    d = 0
    b= 0
    for ch in s:
        if(ch == '/'):
            a.append(d)
        d= d+1
    b= a[-1]
    list.append(s[0: b])
    list.append(s[b+1:])
    return list
s = input('문자열 s =')
print('=>', div_path(s))
```

a라는 리스트를 선언하고 a에 s에서 '/'이 나온 인덱스 위치를 넣고 a의 마지막 값이 마지막 '/'이기에 a의 마지막 값기준으로 짤라준다

# 3. 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
- 집합 자료형 set은 사용하지 않는다.
- 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
- 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
- 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
- 반환하는 리스트는 정렬이 되어 있어야 한다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
- list_union([1, 2, 3], [1, 2, 4])
- [1, 2, 3, 4]
- list_union([-10, -5, 0, -1], [100, 9, 0, 9])
- [-10, -5, -1, 0, 9, 100]
- list_union([0, 1, 2], [0, 1, 2])
- [0, 1, 2]
- [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.
- [참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

```python
def list_union(lista, listb):
    a = []
    d = 0
    c = 0
    b= 0
    for ch in lista:
        for sh in listb:
            if(sh == ch):
                a.append(sh)
            a.append(sh)
        a.append(ch)
    a.sort()
    for ih in a+a:
        if(d<len(a)-1):
            if(a[d] == a[d+1]):
                a.remove(a[d])
                d = d-1
                
        d = d+1
    if(a[-2] == a[-1]):
        a.remove(a[-2]) 
    a.sort()
    return a

listc = [1,2,3]
listd = [1,2,4]
print('list_union(', listc,',', listd, ')')
print(list_union(listc, listd))
```

a라는 리스트를 선언하고 a에다가 lista와 listb를 비교했을 때 같은 값을 넣고 lista와 listb를 넣는다. 그리고 중복된값을 제거한다. 이미 lista와 listb가 같은 값을 들어갔기에 중복을 제거해도 된다. 그리고 중복을 제거할 때 마지막에 인덱스오류가 나므로 예외처리를 했다.

- 모든 가정과 조건은 3번 문제와 동일하다.

```python
def list_intersection(lista, listb):
    a = []
    for ch in lista:
        for sh in listb:
            if(sh == ch):
                a.append(sh)
    a.sort()
    return a
def list_difference(lista,listb):
    a = []
    count = 0
    a = lista + listb
    a.sort()
    listc = a[:]
    for ch in a:
        for sh in a:
            if(ch == sh):
                count = count +1
        if(count>1):
            listc.remove(ch)
        count = 0  
        
    lista.sort()    
    for ih in lista:
        if(count<len(lista)-1):
            if(lista[count] == lista[count+1]):
                listc.append(lista[count])
            count  =count+1
    if(lista[-2] == lista[-1]):
        listc.append(lista[-2])
    count = 0
    listb.sort()    
    for ih in listb:
        if(count<len(listb)-1):
            if(listb[count] == listb[count+1]):
                listc.append(listb[count])
            count  =count+1
    if(listb[-2] == listb[-1]):
        listc.append(listb[-2])
    
    listc.sort()
    return listc

lista = [-10,-5,0,-1]
listb = [100,9,0,9]
print(list_intersection(lista, listb))
print(list_difference(lista,listb))
```

a라는 리스트를 선언하고 a에 lista와 listb를 모두 넣고 listc에 a를 깊은복사를 해줍니다. 그리고 리스트 a를 이중 for문을 통해 검사하여 같은 값이 나오면 count를 1씩 증가하게 만들었습니다. 그러면 lista와 listb의 같은 숫자는 count가 2이상이 될것이고 다른 값은 count는 1이 될것입니다. 이때 count가 2이상이 된 인덱스를 제거하여 차집합을 표현했습니다. 그리고 lista와 listb에 중복된 값이 있을 수 있기 때문에 예외처리를 했습니다.

# 5. 두 개의 양의 정수를 인자로 받아서 해당 범위안에 있는 소수(prime number)의 리스트를 출력하는 함수 print_primenumber(a, b)를 작성하시오.
- 이미 구현된 모듈이나 라이브러리는 사용하지 않는다.
- 함수의 반환(return)은 없으며 함수 내의 끝에서 .format()을 사용하여 다음과 같이 출력한다.
- print_primenumber(2, 10)
- 2 이상 10 이하 정수 중 소수 리스트: [2, 3, 5, 7]

```python
def print_primenumber(a, b):
    list = []
    for c in range(a,b+1):
        result = True
        if(c<2):
            result = False
        for d in range(2,c):
            if(c%d ==0):
                result = False
        if result:
            list.append(c) 
    print("{list}".format(list = list))
print_primenumber(2, 10)
```

정수 a와 b를 받고 리스트를 선언하여 이중 for문과 bool형을 이용해 소수를 리스트에 담아 포멧을 통해 출력했습니다
